```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns  # visualization tool
```

```python
data = pd.read_csv('LinkedIn Profile Data.csv')
data.head()
```

```python
data.info()
```

```python
data.corr()
```

```python
#Heatmap
f,ax = plt.subplots(figsize=(18, 18))
sns.heatmap(data.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
plt.show()
```

```python
# Line Plot

data.	avg_current_position_length.plot(kind = 'line', color = 'g',label = '	avg_current_position_length',linewidth=1,alpha = 0.5,grid = True,linestyle = ':')
data.avg_previous_position_length.plot(color = 'r',label = 'avg_previous_position_length',linewidth=1, alpha = 0.5,grid = True,linestyle = '-.')
plt.legend(loc='upper right')     
plt.xlabel('x axis')              
plt.ylabel('y axis')
plt.title('Line Plot')            
plt.show()
```

```python
# Scatter Plot 
# x =no_of_previous_positions , y = current_position_length
data.plot(kind='scatter', x='no_of_previous_positions', y='current_position_length',alpha = 0.5,color = 'red')
plt.xlabel('no_of_previous_positions')              
plt.ylabel('current_position_length')
plt.title('Scatter Plot')
```

```python
sns.barplot(data['age'].value_counts().index, data['age'].value_counts())
plt.show()
```

```python
#LM plot
sns.lmplot(x='n_followers', y='face_quality', data=data, palette='Set1')
plt.show()
```
